```python
import investpy
import sqlite3
import json
import datetime
 
conn = sqlite3.connect('STOCKS.db')
data = investpy.get_stock_historical_data(stock='AAPL', country='United States', as_json=True, from_date='01/01/2010', to_date='31/12/2020')
#print(data.sort_values(by='Date',ascending=False))
#print(data)
s= json.loads(data)
a= s['historical']

for i in a:
    conn.execute("INSERT INTO STOCKS (DATE,CLOSE) \
      VALUES ("+"'"+str(i['date'])+"'"+","+str(i['close'])+" )");

conn.commit()
print ("Records created successfully");
conn.close()
```
